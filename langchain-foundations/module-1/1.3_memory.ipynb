```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import httpx
from langchain_groq import ChatGroq

# Disable SSL verification to work around corporate proxy/firewall
client = httpx.Client(verify=False)
llm = ChatGroq(model="meta-llama/llama-4-scout-17b-16e-instruct", temperature=0, http_client=client)
```

## No memory

```python
from langchain.agents import create_agent


agent = create_agent(
    llm
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")

response = agent.invoke(
    {"messages": [question]} 
)
```

```python
from pprint import pprint

pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]} 
)

pprint(response)
```

## Memory

```python
from langgraph.checkpoint.memory import InMemorySaver  


agent = create_agent(
    llm,
    checkpointer=InMemorySaver(),  
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is Seán and my favourite colour is green")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages": [question]},
    config,  
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="What's my favourite colour?")

response = agent.invoke(
    {"messages": [question]},
    config,  
)

pprint(response)
```
